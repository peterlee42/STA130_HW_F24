# Analysis 1

## Research Question:
- Does the frequency of engaging in vigorous sports, fitness, or recreational activities per week correlate with an individual's self-esteem regarding their body?

 and does this change between genders

### Variables:

#### EXERCISE_global_physical_activity_questionnaire_leisure_vigorous_days (Independent Variable):
- Ordinal Data
- Measures how many days in a week (1-7) that you do vigorous exercise unrelated to work

#### PSYCH_self_esteem_unknown_scale_body (Dependent Variable):
- Ordinal Data
- Asks respondents to rate their self-esteem about their body, from 1 (not very true) to 7 (very true).

#### Reasoning:
- Physical activity may be a key factor to how an individual feels about their own body.
- I chose vigorous physical activities because I wanted to see if people who are athletic have some sort of correlation with one's own perception of their body.

#### Visualizations:
1. Bar Plot
    - Good to visualize each variable independently since they are ordinal and discrete
    - Will allow us to easily see frequency
2. Side-by-side bar plot
    - May be useful for comparison and may reveal any patterns


### Analysis (Spearman's Rank Correlation):
- Predefined an alpha significance value of 0.05
- I will calculate Spearman's Rank Correlation Coefficient to quantify the strength and direction of the association between these ordinal datasets.
- To do this, we simply assign ranks to each value given by the variables
- If there are repeating values, we give the average rank (ex. d=[3,3,4] would be ranked 1.5,1.5,3 respectively)
- Then we calculate the p-value using an equation to form a conclusion as a form of hypothesis testing
- If it is greater than alpha, then we are unlikely to reject the null hypothesis and if it is less, the we are likely to reject it

#### Assumptions:
- Monotonic Relationship: Assumes a monotonic relationship between the variables.
- Independence: Each observation is assumed to be independent of others.


### Hypothesis:
- Null Hypothesis: There is no significant correlation between the number of days of vigorous exercise and self esteem about one's body.
- Alternative Hypothesis: There is a significant correlation between the number of days of vigorous exercise and self esteem about one's body.
- I hypothesis that there will be a positive correlation since it is often understood that people who exercise typically have healthier bodies which may improve self esteem. 

##### Possible Results
- P-value <= 0.05 rejecting null hypothesis
- P-value > 0.05 failing to reject null hypothesis

### Relevance
- We are able to understand whether or not vigorously exercising is overall better for individual's self esteem
- If there is no correlation, there could be other factors beyond exercising that influences one's self perception of their body.

### Analysis (Fisher's Exact Test):
1. Determine a alpha significance value of 0.05
2. Construct a 4x3 contingency table such that each cell will contain the count of individuals that fall into the paired combination of each category
3. Then use Fisher's Exact Test to compute a significant p-value (this will be more detailed later)
4. Compare the computed p-value to the alpha significance value to determine if we reject or fail to reject the null hypothesis

#### Assumptions
- Data should be representative of the population
- Each response is independent
- Categories are mutually exclusive
