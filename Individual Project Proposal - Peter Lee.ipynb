# Analysis 2:
### Research Question:
-  Does the number of close friends correspond to feeling left out?
### Variables:
#### CONNECTION_social_num_close_friends_grouped (Independent)
- Represents the number of close friends one has, grouped into the categories: None, 1-2, 3-4, 5 or more
#### LONELY_ucla_loneliness_scale_left_out (Dependent)
- Asks how often you feel left out with the options: Some of the time, Often, Hardly Ever
##### Reasoning:
- Often, it seems as though having more friends will always make you feel welcomed, but I wonder if having a greater number of close friends ends up making it more difficult to have a closer connection, leading to feeling left out.
#### Visualizations:
Bar Plot
- Good to visualize each variable independently since they are ordinal and discrete
- Will allow us to easily see frequency
### Analysis:
1. Determine a alpha significance value of 0.05
2. Construct a 4x3 contingency table such that each cell will contain the count of individuals that fall into the paired combination of each category
3. Use Fisher's Exact Test to compute a significant p-value (using scipy.stats)
4. Compare the computed p-value to the alpha significance value to determine if we reject or fail to reject the null hypothesis
5. Maybe look into permutation testing to find more interesting facts aobut the correlation if any exist.
#### Assumptions
- Data should be representative of the population
- Each response is independent
- Categories are mutually exclusive
### Hypothesis:
- H0: There is no relationship between the number of close friends and the frequency of feeling left out
- HA: There is an association between the number of close friends and the frequency of feeling left out
- I expect to see a correlation between the number of friends and feeling left out. since more connections means more opportunities to engage in activities.
##### Possible Results
- P-value <= 0.05 rejecting null hypothesis
- P-value > 0.05 failing to reject null hypothesis
### Relevance:
- This analysis can show insights to whether the number of close friends is a great factor to feeling left out by looking at the correlation.
